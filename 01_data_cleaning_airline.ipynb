```python

# for tprice
# ============================================================
# PART 1: NON-ZERO DISTANCE PRICE FILL (AADT DICTIONARY)
# ============================================================

def nrst_to_mean(li):
    if len(li) == 0:
        return np.nan
    me = np.mean(li)
    nrs = min(range(len(li)), key=lambda x: abs(li[x] - me))
    return li[nrs]

# AADT: Airline → Aircraft → Distance → Ticket_Price
AADT = {}

for airline in df["Airline"].unique():
    AADT[airline] = {}
    ac_types = df[df["Airline"] == airline]["Aircraft_Type"].unique()

    for ac in ac_types:
        AADT[airline][ac] = {}
        distances = (
            df[(df["Airline"] == airline)
               & (df["Aircraft_Type"] == ac)
               & (df["Distance_km"] != 0)]
            ["Distance_km"]
            .unique()
            .tolist()
        )

        for dist in distances:
            prices = (
                df[(df["Airline"] == airline)
                   & (df["Aircraft_Type"] == ac)
                   & (df["Distance_km"] == dist)
                   & (df["Ticket_Price"].notna())]
                ["Ticket_Price"]
                .unique()
                .tolist()
            )
            AADT[airline][ac][dist] = nrst_to_mean(prices)

# ---- Fill non-zero distances ----
for i, j in AADT.items():
    for m, n in j.items():
        for q, w in n.items():
            df.loc[
                (df["Airline"] == i)
                & (df["Aircraft_Type"] == m)
                & (df["Distance_km"] == q)
                & (df["Ticket_Price"].isna()),
                "Ticket_Price"
            ] = w


# ============================================================
# PART 2: BUILD OD (Origin → Destination → Distance)
# FOR ZERO-DISTANCE ENTRIES
# ============================================================

OD = {}
for o in df["Origin"].unique():
    OD[o] = {}
    dests = df[(df["Origin"] == o) & (df["Origin"] != df["Destination"])]["Destination"].unique()
    for d in dests:
        dist_list = (
            df[(df["Origin"] == o)
               & (df["Destination"] == d)
               & (df["Distance_km"] != 0)]
            ["Distance_km"]
            .unique()
            .tolist()
        )
        if len(dist_list) > 0:
            OD[o][d] = dist_list[0]


# ============================================================
# PART 3: FILL ZERO-DISTANCE PRICE (Origin == Destination)
# ============================================================

for airline, ac_dict in AADT.items():
    for ac, dist_dict in ac_dict.items():
        for origin, dest_dict in OD.items():

            # random destination for this origin
            if len(dest_dict) == 0:
                continue

            des = np.random.choice(list(dest_dict.keys()))
            dis = dest_dict[des]

            if dis not in dist_dict:
                continue

            price_val = dist_dict[dis]

            df.loc[
                (df["Airline"] == airline)
                & (df["Aircraft_Type"] == ac)
                & (df["Origin"] == origin)
                & (df["Origin"] == df["Destination"])    # zero-distance condition
                & (df["Ticket_Price"].isna()),
                "Ticket_Price"
            ] = price_val


# ============================================================
# PART 4: FINAL CLEANUP — REMOVE ALL REMAINING NULL VALUES
# ============================================================

# 1. Fill missing using airline mean
df["Ticket_Price"] = df["Ticket_Price"].fillna(
    df.groupby("Airline")["Ticket_Price"].transform("mean")
)

# 2. Fill any last remaining with global mean
df["Ticket_Price"] = df["Ticket_Price"].fillna(df["Ticket_Price"].mean())

# ============================================================
# DONE — 0 NULL VALUES WILL REMAIN
# ============================================================

print("Remaining NULL Ticket_Price values:", df["Ticket_Price"].isna().sum())
```

```python
#for distance_km
li = df[(df["Distance_km"] == 0.0) & (df["Distance_km"].notna())]["Distance_km"].unique().tolist()

dic = {}

for i in li:
    dic[i] = [
        df[df["Distance_km"] == i]["Origin"].unique().tolist(),
        df[df["Distance_km"] == i]["Destination"].unique().tolist()
    ]

# 0.0 values
df.loc[
    (df["Origin"] == df["Destination"]) & (df["Distance_km"].isna()),
    "Distance_km"
] = 0.0

# rest other values
for i, j in dic.items():
    df.loc[
        (df["Origin"].isin(j[0])) &
        (df["Destination"].isin(j[1])) &
        (df["Distance_km"].isna()),
        "Distance_km"
    ] = i
    
```

```python
import numpy as np


def nrst_to_mean(li):
    """Return the value closest to the mean."""
    if len(li) == 0:
        return np.nan
    me = np.mean(li)
    nrs = min(range(len(li)), key=lambda x: abs(li[x] - me))
    return li[nrs]




AADT = {
    i: {
        p: {
            q: nrst_to_mean(
                df[
                    (df["Airline"] == i)
                    & (df["Aircraft_Type"] == p)
                    & (df["Distance_km"] == q)
                    & (df["Ticket_Price"].notna())
                ]["Ticket_Price"].unique().tolist()
            )
            for q in o
        }
        for p, o in {
            k: df[
                (df["Airline"] == i)
                & (df["Aircraft_Type"] == k)
                & (df["Distance_km"] != 0.0)
            ]["Distance_km"].unique().tolist()
            for k in j
        }.items()
    }
    for i, j in {
        i: df[df["Airline"] == i]["Aircraft_Type"].unique().tolist()
        for i in df["Airline"].unique().tolist()       # FIXED (.items removed)
    }.items()
}




for i, j in AADT.items():
    for m, n in j.items():
        for q, w in n.items():
            if pd.isna(w):  # skip invalid fills
                continue
            df.loc[
                (df["Airline"] == i)
                & (df["Aircraft_Type"] == m)
                & (df["Distance_km"] == q)
                & (df["Ticket_Price"].isna()),
                "Ticket_Price"
            ] = w




OD = {}

for origin in df["Origin"].unique().tolist():
    OD[origin] = {}
    destinations = (
        df[(df["Origin"] == origin) & (df["Origin"] != df["Destination"])]["Destination"]
        .unique()
        .tolist()
    )

    for dest in destinations:
        dist_list = (
            df[(df["Origin"] == origin) & (df["Destination"] == dest)]
            ["Distance_km"]
            .unique()
            .tolist()
        )

       
        if len(dist_list) == 0:
            continue

        OD[origin][dest] = dist_list[0]




for i, j in AADT.items():          # Airline → dict
    for m, n in j.items():         # Aircraft → dict
        for o, d in OD.items():    # Origin → {destinations}
            if len(d) == 0:
                continue

            des = np.random.choice(list(d.keys()))
            dis = d[des]

            if dis not in n:       # skip missing keys
                continue

            df.loc[
                (df["Airline"] == i)
                & (df["Aircraft_Type"] == m)
                & (df["Origin"] == o)
                & (df["Origin"] == df["Destination"])  # distance = 0
                & (df["Ticket_Price"].isna()),
                "Ticket_Price"
            ] = n[dis]
```

```python
def dhf(a):
    if isinstance(a, str):
        return a   # already in HH:MM:SS format
    
    h = int(a)
    m = int((a - h) * 60)
    s = round(((a - h) * 60 - m) * 60)

    if s >= 60:
        m += s // 60
        s = s % 60
    if m >= 60:
        h += m // 60
        m = m % 60

    return f"{h:02d}:{m:02d}:{s:02d}"

df["Duration_hours"] = df["Duration_hours"].apply(dhf)
```

```python
def dmf(a):
    if isinstance(a, str):
        return a   # already formatted like '000:000'
    
    h = 0
    m = int(a)
    s = round((a - m) * 60)

    if s >= 60:
        m += s // 60
        s = s % 60
    if m >= 60:
        h += m // 60
        m = m % 60

    return f"{h:02d}:{m:02d}:{s:02d}"

df["Delay_minutes"] = df["Delay_minutes"].apply(dmf)
```
