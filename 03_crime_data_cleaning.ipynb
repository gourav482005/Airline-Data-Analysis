```python
# for district
# Step 1: Street-wise most common District (MODE)
street_district_mode = (
    df.dropna(subset=["STREET", "DISTRICT"])
      .groupby("STREET")["DISTRICT"]
      .agg(lambda x: x.mode().iloc[0])
)

# Step 2: Fill missing District using Street
df.loc[
    df["DISTRICT"].isna() & df["STREET"].notna(),
    "DISTRICT"
] = df.loc[
    df["DISTRICT"].isna() & df["STREET"].notna(),
    "STREET"
].map(street_district_mode)
ra_district_mode = (
    df.dropna(subset=["REPORTING_AREA", "DISTRICT"])
      .groupby("REPORTING_AREA")["DISTRICT"]
      .agg(lambda x: x.mode().iloc[0])
)

df.loc[
    df["DISTRICT"].isna(),
    "DISTRICT"
] = df.loc[
    df["DISTRICT"].isna(),
    "REPORTING_AREA"
].map(ra_district_mode)
```

```python
ra_district_mode = (
    df.dropna(subset=["REPORTING_AREA", "DISTRICT"])
      .groupby("REPORTING_AREA")["DISTRICT"]
      .agg(lambda x: x.mode().iloc[0])
)

df.loc[
    df["DISTRICT"].isna(),
    "DISTRICT"
] = df.loc[
    df["DISTRICT"].isna(),
    "REPORTING_AREA"
].map(ra_district_mode)
```

```python
import numpy as np

# 1️⃣ Mode STREET for each (DISTRICT + REPORTING_AREA)
street_map = (
    df.dropna(subset=["STREET"])
      .groupby(["DISTRICT", "REPORTING_AREA"])["STREET"]
      .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
)

# 2️⃣ Create Series aligned with df rows
street_fill = df.apply(
    lambda r: street_map.get((r["DISTRICT"], r["REPORTING_AREA"]), np.nan),
    axis=1
)

# 3️⃣ Fill STREET using that Series
df["STREET"] = df["STREET"].fillna(street_fill)

# 4️⃣ Fallback: DISTRICT-level mode
district_mode = (
    df.dropna(subset=["STREET"])
      .groupby("DISTRICT")["STREET"]
      .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
)

df["STREET"] = df["STREET"].fillna(df["DISTRICT"].map(district_mode))
```

```python
import numpy as np
import pandas as pd

# 1️⃣ DISTRICT + STREET → mode Lat & Long
ds_map = (
    df.dropna(subset=["Lat", "Long"])
      .groupby(["DISTRICT", "STREET"])[["Lat", "Long"]]
      .agg(lambda x: x.mode().iloc[0])
)

# 2️⃣ STREET → mode Lat & Long
street_map = (
    df.dropna(subset=["Lat", "Long"])
      .groupby("STREET")[["Lat", "Long"]]
      .agg(lambda x: x.mode().iloc[0])
)

# 3️⃣ DISTRICT → mode Lat & Long
district_map = (
    df.dropna(subset=["Lat", "Long"])
      .groupby("DISTRICT")[["Lat", "Long"]]
      .agg(lambda x: x.mode().iloc[0])
)

# 4️⃣ Fill Lat & Long safely
for idx, row in df.loc[df["Lat"].isna() | df["Long"].isna()].iterrows():

    if (row["DISTRICT"], row["STREET"]) in ds_map.index:
        df.loc[idx, "Lat"]  = ds_map.loc[(row["DISTRICT"], row["STREET"]), "Lat"]
        df.loc[idx, "Long"] = ds_map.loc[(row["DISTRICT"], row["STREET"]), "Long"]

    elif row["STREET"] in street_map.index:
        df.loc[idx, "Lat"]  = street_map.loc[row["STREET"], "Lat"]
        df.loc[idx, "Long"] = street_map.loc[row["STREET"], "Long"]

    elif row["DISTRICT"] in district_map.index:
        df.loc[idx, "Lat"]  = district_map.loc[row["DISTRICT"], "Lat"]
        df.loc[idx, "Long"] = district_map.loc[row["DISTRICT"], "Long"]
```
